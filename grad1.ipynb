```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.arange(50)
y_train = 0.1*x_train + np.sin(x_train)
plt.scatter(x_train,y_train)
m = len(x_train)
```

```python
#compute cost function
def compute_cost(x, y, theta_0, theta_1):
    cost_sum = 0
    m = len(x)
    for i in range (m):
        h = theta_0 + theta_1*x[i]
        cost = (h - y[i])**2
        cost_sum = cost_sum + cost
    total_cost = 0.5*cost_sum
    return total_cost
```

```python
theta_0 = 0
theta_1 = 1
cost = compute_cost(x_train, y_train, theta_0, theta_1)
print(cost)
```

```python
#compute gradients
def gradient(x, y, theta_0, theta_1):
    m = len(x)
    dj_dt0 = 0
    dj_dt1 = 0
    for i in range (m):
        h = theta_0 + theta_1*x[i]
        dj_dt0_f = (h - y[i])
        dj_dt1_f = (h- y[i])*x[i]
        
        dj_dt0 = dj_dt0 + dj_dt0_f
        dj_dt1 = dj_dt1 + dj_dt1_f
    return dj_dt0, dj_dt1
```

```python
dj_dt0, dj_dt1 = gradient(x_train, y_train, theta_0, theta_1)
print(dj_dt0, dj_dt1)
```

```python
#compute theta_0 and theta_1
def gradient_descent(x, y, theta_0, theta_1, alpha, num, gradient):

    theta_0 = theta_0
    theta_1 = theta_1
    
    for i in range (num):    
        dj_dt0, dj_dt1 = gradient(x, y, theta_0, theta_1)
        
        theta_0 = theta_0 - alpha*dj_dt0
        theta_1 = theta_1 - alpha*dj_dt1
        
    return theta_0, theta_1
```

```python
theta_0 = 0
theta_1 = 0
alpha = 0.00001
num = 2000
final_0, final_1 = gradient_descent(x_train, y_train, theta_0, theta_1, alpha, num, gradient)
print(final_0, final_1)
```

```python
plt.plot(x_train, final_0 + final_1*x_train, 'b')
plt.scatter(x_train, y_train)
```
