Implementing Least Squares to determine the parameters

```python
import numpy as np
import matplotlib.pyplot as plt

#inverse calculating method
from numpy.linalg import inv
```

```python
#Data generation
x_train = np.arange(50)  
y_train = 0.1*x_train + (x_train) ** 0.5
ones = np.ones(50)
plt.scatter(x_train, y_train)
```

```python
#Design matrix
X = np.stack((ones, x_train), axis = 1)

#Output matrix
Y = y_train
```

```python
#theta = inv(tr(X)(X))tr(X)Y

x_tr = X.transpose()
xtx = np.matmul(x_tr,X)
inverse = inv(xtx)
ixtxxt = np.matmul(inverse, x_tr)
theta = np.matmul(ixtxxt, Y)
```

```python
plt.plot(x_train, theta[1]*x_train + theta[0])
plt.scatter(x_train, y_train, marker='x', c = 'r')
```
