Gradient Descent Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1, 2])
y_train = np.array([300, 500])
```

```python
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
plt.scatter(x_train, y_train, marker="x", c="r")
plt.title("housing prices")
plt.xlabel("1000 square feet")
plt.ylabel("price in  1000 dollars")
```

```python
m = len(x_train)
print(f"number of training example = {m}")
```

```python
print(f"(x(0), y(0)) = {x_train[0], y_train[0]}")
```

```python
#computing f_wb

w = 0
b = 0

def compute_model_output(x, w, b):
    m = len(x_train)
    f_wb = np.zeros(m)
    for i in range (m):
        f_wb[i] = w*x[i] + b

    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, tmp_f_wb, c='b', label="our expectation")
plt.scatter(x_train, y_train, marker='x', c='r', label="actual values")
plt.title("housing prices")
plt.xlabel("1000 square feet")
plt.ylabel("price in  1000 dollars")
plt.legend()
plt.show()
```

```python
#computing cost function
def compute_cost (x, y, w, b):
    cost_sum = 0
    for i in range (m):
        f_wb = w*x[i] + b
        cost = (f_wb - y[i])**2
        cost_sum = cost_sum + cost
    total_cost = (1/(2*m))*cost_sum
    
    return total_cost
```

```python
r = compute_cost(x_train, y_train, w, b)
print(f"total cost = {r}")
```

```python
#Compute gradients dj/dw and dj/db
def gradient(x, y, w, b):
    sum_f_dw = 0
    sum_f_db = 0
    for i in range (m):
        f_wb = w*x[i] + b
        sum_db = (f_wb - y[i])
        sum_dw = sum_db*x[i]
        sum_f_dw = sum_f_dw + sum_dw
        sum_f_db = sum_f_db + sum_db
    dj_dw = (1/m)*sum_f_dw
    dj_db = (1/m)*sum_f_db

    return dj_dw, dj_db
```

```python
dj_dw, dj_db = gradient(x_train, y_train, w, b)
print(dj_dw, dj_db)
```

```python
#compute w and b using gradient descent
def gradient_descent(x, y, w, b, num, alpha, gradient):
    
    b = b
    w = w
    
    for i in range(num):
        dj_dw, dj_db = gradient(x, y, w , b)     

        b = b - alpha * dj_db                            
        w = w - alpha * dj_dw                            

    return w, b
```

```python
w = 0
b = 0
num = 1000
alpha = 0.01
w_final, b_final = gradient_descent(x_train, y_train, w, b, num, alpha, gradient)
print(w_final, b_final)
```

```python
plt.plot(x_train, x_train*w_final + b_final, 'r')
plt.scatter(x_train, y_train, marker='+', c='b')
```
